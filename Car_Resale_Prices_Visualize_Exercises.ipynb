```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
color= sns.color_palette("pastel")
sns.set(style="whitegrid", context="talk", font_scale=0.8, palette=color,rc={'figure.figsize':(8,4)})
pd.set_option('display.float_format', lambda x: '%.f' % x)
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/egecandrsn/Car_resale_data-2023/main/cleaned.csv", index_col=0)
```

```python
data.head()
```

```python
#Create a histogram or density plot of the resale_price column.

sns.kdeplot(data=data, x="resale_price",fill=True)
plt.show()
```

```python
#Create a box plot or violin plot of resale_price for each fuel_type.

sns.boxplot(data=data,x="fuel_type",y="resale_price")
plt.show()
```

```python
#Create a scatter plot of registered_year vs. resale_price.

sns.regplot(data=data, x="registered_year",y="resale_price", line_kws={"color":color[1]})
```

```python
#Create a bar plot or countplot of body_type to visualize the distribution.

sns.countplot(data=data,x="body_type",order=["Hatchback","Sedan","SUV","MUV"])
plt.show()
```

```python
#Create a scatter plot of engine_capacity vs. mileage and observe any patterns.

sns.regplot(data=data,x="engine_capacity",y="mileage", line_kws={"color":color[1]})
plt.show()
```

```python
#Create a countplot or histogram of the seats column.
sns.countplot(data=data, x="seats", width=0.5)
plt.show()
```

```python
#Create a countplot to visualize the distribution of transmission_type_Automatic and transmission_type_Manual.

sns.countplot(data=data, x="transmission_type_Manual")
plt.show()
```

```python
#Create a scatter plot of kms_driven vs. resale_price.

sns.regplot(data=data, x="kms_driven", y="resale_price",line_kws={"color":color[1]})
plt.show()
```

```python
#Create a box plot or violin plot of resale_price for each owner_type.

plt.figure(figsize=(8,10))
sns.boxplot(data=data, x="owner_type",y="resale_price")
plt.show()
```

```python
#Create a scatter plot of engine_capacity vs. mileage and calculate the correlation coefficient.

import scipy
r = scipy.stats.pearsonr(x=data["engine_capacity"], y=data["mileage"])[0]
sns.regplot(data=data, x="engine_capacity",y="mileage",line_kws={"color":color[1]})
plt.text(100, 3100, 'r = ' + str(round(r, 3)), fontsize=25)
plt.show()
```

```python
#Create a line plot of the count of cars registered per year

sns.lineplot(x=data.registered_year.value_counts().to_dict().keys(),y=data.registered_year.value_counts().to_dict().values())
plt.show()
```

```python
#Create a scatter plot matrix with kms_driven, registered_year, and fuel_type, and use colors or markers to distinguish the fuel types.

sns.pairplot(data=data[["kms_driven", "registered_year", "fuel_type"]], hue="fuel_type", kind="scatter")
plt.show()
```

```python
#Create a multi-category box plot or violin plot to explore the joint effect of fuel_type and transmission_type on resale_price.

sns.boxplot(data=data, x="fuel_type", y="resale_price", hue="transmission_type_Manual")
plt.show()
```

```python
#Create a grouped box plot or violin plot to visualize resale_price for different engine_capacity ranges within each fuel_type category.

engine_capacity_group= pd.cut(data["engine_capacity"], bins=range(500, int(data["engine_capacity"].max()) + 500, 500))

sns.boxplot(data=data,x=engine_capacity_group,y="resale_price", hue="fuel_type")
plt.show()
```

```python
#Create density plot to investigate the relationship between registered_year, kms_driven, and fuel_type.

sns.kdeplot(data=data,x="kms_driven",y="registered_year",hue="fuel_type")
plt.show()
```

```python
#Create a countplot to visualize the frequency of each unique combination of body_type and fuel_type.


sns.countplot(data=data, x="fuel_type")
plt.show()
sns.countplot(data=data, x="body_type")
plt.show()
```

```python
#Create a line plot or time series plot to visualize the trend in car registrations over time, segmented by different cities.

plt.figure(figsize=(15,6))
city_registration_counts = data.groupby(['registered_year', 'city']).size().unstack().fillna(0)
sns.lineplot(data=city_registration_counts, dashes=False)
plt.show()
```

```python
#Create scatter plots or regression plots to investigate the relationship between kms_driven, resale_price, and fuel_type.


sns.scatterplot(data=data, x='kms_driven', y='resale_price', hue='fuel_type')
sns.regplot(data=data, x='kms_driven', y='resale_price', scatter=False, color='gray')
plt.show()
```

```python
#Create box plots to visualize resale_price for each combination of insurance type and body_type.

sns.boxplot(data=data,x="insurance",y="resale_price",hue="body_type")
plt.show()
```
